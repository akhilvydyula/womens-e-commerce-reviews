# Women's E-Commerce Reviews: End-to-End ML Walkthrough

Learning path:
1. Load and inspect data
2. Baseline model
3. Better model (text + tabular)
4. Advanced model + tuning idea
5. Submission file creation


````python
import pandas as pd
from src.config import RAW_FILE_PATH
from src.data import basic_cleaning

df = pd.read_csv(RAW_FILE_PATH)
df = basic_cleaning(df)
print(df.shape)
df.head()
````

````python
from src.data import make_text_feature, split_data
from src.train import get_model_pipeline, evaluate_binary

work = df.copy()
work['text'] = make_text_feature(work)

X_train, X_valid, y_train, y_valid = split_data(work)
pipeline = get_model_pipeline('baseline')
pipeline.fit(X_train, y_train)

pred = pipeline.predict(X_valid)
proba = pipeline.predict_proba(X_valid)[:, 1]
metrics = evaluate_binary(y_valid, pred, proba)
metrics
````

````python
# Try better model
pipeline = get_model_pipeline('better')
pipeline.fit(X_train, y_train)
pred = pipeline.predict(X_valid)
proba = pipeline.predict_proba(X_valid)[:, 1]
evaluate_binary(y_valid, pred, proba)
````
